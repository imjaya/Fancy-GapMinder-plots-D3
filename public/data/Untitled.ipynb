```python
import pandas as pd
import numpy as np
```

```python
income=pd.read_csv("income_new.csv")
life_expectancy=pd.read_csv("life_expectancy_new.csv")
population=pd.read_csv("population_new.csv")
emmision=pd.read_csv("emmision_new.csv")
countries=pd.read_csv("countries_regions.csv")
mortality=pd.read_csv("child_mortality_0_5_year_olds_dying_per_1000_born.csv")
```

```python
result=[]
```

```python
for i in countries['name']:
    x={
        "country":i,
        "code":countries['geo'][countries[countries['name'] == str(i)].index[0]],
        "region":countries['World bank region'][countries[countries['name'] == str(i)].index[0]],
        "income":[],
        "life_expectancy":[],
        "emmision":[],
        "population":[],
        "mortality":[]
        }
    
    for i in range(1800,2101):
        #income
        if(str(i) in list(income.columns)):
            if(income[str(i)][income[income['country'] == x['country']].index.tolist()]==None):
                x["income"].append(0)
                
            else:
                x["income"].append(float(income[str(i)][income[income['country'] == x['country']].index.tolist()]))
        else:
            x["income"].append(0)
            
        #life_expectancy    
        if(str(i) in list(income.columns)):
            if(life_expectancy[str(i)][life_expectancy[life_expectancy['country'] == x['country']].index.tolist()]==nan):
                x["life_expectancy"].append(0)
            else:
                x["life_expectancy"].append(float(life_expectancy[str(i)][life_expectancy[life_expectancy['country'] == x['country']].index.tolist()]))
        else:
            x["life_expectancy"].append(0)
            
        #population    
        if(str(i) in list(income.columns)):
            if(population[str(i)][population[population['country'] == x['country']].index.tolist()]==nan):
                x["population"].append(0)
                
            else:
                x["population"].append(float(population[str(i)][population[population['country'] == x['country']].index.tolist()]))
        else:
            x["population"].append(0)
        
        #emmision    
        if(str(i) in list(income.columns)):
            if(emmision[str(i)][emmision[emmision['country'] == x['country']].index.tolist()]==nan):
                x["emmision"].append(0)
            else:       
                x["emmision"].append(float(emmision[str(i)][emmision[emmision['country'] == x['country']].index.tolist()]))
        else:
            x["emmision"].append(0)
            
            
        #emmision    
        if(str(i) in list(income.columns)):
            if(mortality[str(i)][mortality[mortality['country'] == x['country']].index.tolist()]==nan):
                x["mortality"].append(0)
            else:       
                x["mortality"].append(float(mortality[str(i)][mortality[mortality['country'] == x['country']].index.tolist()]))
        else:
            x["mortality"].append(0)
    print(x)
    result.append(x)
```

```python
emmision.drop('Eswatini', inplace=True, axis=1)
emmision.drop('North Macedonia', inplace=True, axis=1)
population.drop('Eswatini', inplace=True, axis=1)
population.drop('North Macedonia', inplace=True, axis=1)
life_expectancy.drop('Eswatini', inplace=True, axis=1)
life_expectancy.drop('North Macedonia', inplace=True, axis=1)
income.drop('Eswatini', inplace=True, axis=1)
income.drop('North Macedonia', inplace=True, axis=1)
mortality.drop('Eswatini', inplace=True, axis=1)
mortality.drop('North Macedonia', inplace=True, axis=1)
```
